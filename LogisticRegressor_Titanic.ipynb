```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
print('train data shape:',train.shape)
print('test data shape:',test.shape)
```

```python
full = pd.concat([train,test],sort=False,ignore_index=True)
print('all data shape:',full.shape)
```

```python
full.describe()
```

```python
full.info()
```

```python
full.head()
```

```python
full.tail()
```

```python
full['Age'] = full['Age'].fillna(full['Age'].mean())
```

```python
full.info()
```

```python
full['Fare'] = full['Fare'].fillna(full['Fare'].mean())
```

```python
full.info()
```

```python
full['Embarked'].head()
```

```python
full['Embarked'].value_counts()
```

```python
full['Embarked'] = full['Embarked'].fillna('S')
```

```python
full.info()
```

```python
full['Cabin'].head()
```

```python
full['Cabin'] = full['Cabin'].fillna('U')
```

```python
full.info()
```

```python
full.head()
```

```python
sexDic = {'male':1,'female':0}
full['Sex'] = full['Sex'].map(sexDic)
full.head()
```

```python
full['Pclass'].value_counts()
```

```python
pclassDf = pd.DataFrame()
pclassDf = pd.get_dummies(full['Pclass'],prefix='Pclass')
pclassDf.head()
```

```python
full = pd.concat([full,pclassDf],axis=1)
full.head()
```

```python
full.drop('Pclass',axis=1,inplace=True)
full.head()
```

```python
embarkedDf = pd.DataFrame()
embarkedDf = pd.get_dummies(full['Embarked'],prefix='Embarked')
embarkedDf.head()
```

```python
full = pd.concat([full,embarkedDf],axis=1)
full.head()
```

```python
full.drop('Embarked',axis=1,inplace=True)
full.head()
```

```python
full['Cabin'] = full['Cabin'].map(lambda x:x[0])
full.head()
```

```python
cabinDf = pd.DataFrame()
cabinDf = pd.get_dummies(full['Cabin'],prefix='Cabin')
cabinDf.head()
```

```python
full = pd.concat([full,cabinDf],axis=1)
full.head()
```

```python
full.drop('Cabin',axis=1,inplace=True)
full.head()
```

```python
def get_title(name):
    return name.split(',')[1].split('.')[0].strip()
```

```python
get_title('Braund, Mr. Owen Harris')
```

```python
full['Name'] = full['Name'].map(get_title)
full.head()
```

```python
titleDic = {
                    "Capt":       "Officer",
                    "Col":        "Officer",
                    "Major":      "Officer",
                    "Jonkheer":   "Royalty",
                    "Don":        "Royalty",
                    "Sir" :       "Royalty",
                    "Dr":         "Officer",
                    "Rev":        "Officer",
                    "the Countess":"Royalty",
                    "Dona":       "Royalty",
                    "Mme":        "Mrs",
                    "Mlle":       "Miss",
                    "Ms":         "Mrs",
                    "Mr" :        "Mr",
                    "Mrs" :       "Mrs",
                    "Miss" :      "Miss",
                    "Master" :    "Master",
                    "Lady" :      "Royalty"
}
```

```python
titleDf = pd.DataFrame()
titleDf = full['Name'].map(titleDic)
titleDf.head()
```

```python
titleDf.value_counts()
```

```python
titleDf = pd.get_dummies(titleDf)
```

```python
type(titleDf)
```

```python
titleDf.head()
```

```python
full = pd.concat([full,titleDf],axis=1)
full.head()
```

```python
full.drop('Name',axis=1,inplace=True)
```

```python
full.head()
```

```python
familyDf = pd.DataFrame()
familyDf['FamilySize'] = full['SibSp'] + full['Parch'] + 1
familyDf.head()
```

```python
familyDf['FamilySingle'] = familyDf['FamilySize'].map(lambda x:1 if x==1 else 0)
familyDf['FamilySmall'] = familyDf['FamilySize'].map(lambda x:1 if 2<=x<=4 else 0)
familyDf['FamilyLarge'] = familyDf['FamilySize'].map(lambda x:1 if x>=5 else 0)
familyDf.head()
```

```python
full = pd.concat([full,familyDf],axis=1)
full.shape
```

```python
full.corr()
```

```python
full.corr()['Survived'].sort_values(ascending=False)
```

```python
X_feature = pd.DataFrame()
```

```python
X_feature = pd.concat([titleDf,pclassDf,familyDf,full['Fare'],cabinDf,embarkedDf,full['Sex']],axis=1)
X_feature.shape
```

```python
train_data = X_feature.loc[0:890,:]
train_result = full.loc[0:890,'Survived']
pred_data = X_feature.loc[891:,:]
```

```python
X_train,X_test,y_train,y_test = train_test_split(train_data,train_result,random_state=14,train_size=0.8)
```

```python
print('X_train shape:',X_train.shape)
print('X_test shape:',X_test.shape)
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
pred_result = model.predict(pred_data)
pred_result.shape
```

```python
pred_result[0:5]
```

```python
pred_resultDf = pd.DataFrame({'PassengerId':test['PassengerId'],'Survived':pred_result})
```

```python
pred_resultDf.head()
```

```python
pred_resultDf['Survived'] = pred_resultDf['Survived'].astype(int)
pred_resultDf.head()
```

```python
pred_resultDf.to_csv('predic_result.csv',index=False)
```
